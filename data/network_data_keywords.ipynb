```python
import pandas as pd 
import matplotlib.pyplot as plt

network_df= pd.read_csv("Networking-Info-csv-txt.csv")
```

```python
keywords_columns = ['KEYWORDS 1', 'KEYWORDS 2', 'KEYWORDS 3', 'KEYWORDS 4', 'KEYWORDS 5', 'KEYWORDS 6'] 
keywords = network_df[keywords_columns].stack().dropna()

keyword_frequencies = keywords.value_counts()


top_keywords = keyword_frequencies.head(15)

plt.figure(figsize=(10, 6))
top_keywords.plot(kind='bar', color='steelblue')
plt.title('Top 15 Most Common Keywords')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
